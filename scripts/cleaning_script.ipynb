```python
import pandas as pd
```

```python
df = pd.read_csv("TSLA_stock.csv")
```

```python
df = df.replace(' ','-',regex=True)
df
```

```python
replaced_values = {"Jan":"01","Feb":"02","March":"03","Apr":"04","May":"05","Jun":"06","Jul":"07","Aug":"08","Sep":"09","Oct":"10","Nov":"11","Dec":"12"}
df = df.replace({"Date":replaced_values},regex=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.to_csv("/data/TSLA_stock_cleaned.csv",index=False)
```
